```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot (x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
values = [10, 20, 15, 25]

# Create a bar chart
plt.bar(categories, values)

# Add labels
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Show the chart
plt.show()
```

```python
from matplotlib import pyplot as plt

year =  [2050,2055,2060,2065,2070,2075,2080]
index = [300,250,360,900,500,1600,1850]

plt.plot(year, index, color='green')
#plt.bar(year, index)
plt.scatter(year, index, color='green')

plt.xlabel('Year')
plt.ylabel('Index')
plt.title('Nepse Index')
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'rD')
plt.show()
```
